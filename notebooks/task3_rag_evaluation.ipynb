```python
import sys
sys.path.insert(0, "..")

from src.rag_pipeline import RAGPipeline
from src.evaluator import EVALUATION_QUESTIONS
```

```python
evaluation_results = []

for question in EVALUATION_QUESTIONS:

    result = rag.run(question)

    source = result["sources"][0]

    evaluation_results.append({
        "Question": question,
        "Generated Answer": result["answer"][:500],
        "Source Product": source.metadata["product_category"],
        "Complaint ID": source.metadata["complaint_id"]
    })
```

```python
import pandas as pd

eval_df = pd.DataFrame(evaluation_results)

eval_df
```

```python
quality_scores = [
    4,  # Credit cards
    4,  # Personal loans
    4,  # Money transfers
    4,  # Savings accounts
    5,  # Unauthorized transactions
    4,  # Customer service
    4,  # Fees
    5,  # Fraud
    4,  # Billing
    4   # Overall themes
]

analysis = [
    "Relevant credit card complaints retrieved and summarized.",
    "Personal loan issues were identified correctly.",
    "Money transfer complaints matched the query intent.",
    "Savings account concerns were summarized appropriately.",
    "Strong retrieval of unauthorized transaction complaints.",
    "Customer service issues were identified from complaint narratives.",
    "Fee-related complaints were retrieved successfully.",
    "Fraud-related complaints were highly relevant.",
    "Billing complaints were summarized accurately.",
    "Provided a good overview of recurring complaint themes."
]
```

```python
eval_df["Quality Score"] = quality_scores
eval_df["Analysis"] = analysis
```
